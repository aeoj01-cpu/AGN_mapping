# BLR Reverberation Mapping Simulation

This notebook simulates broad-line region (BLR) reverberation by convolving an AGN continuum light curve with a BLR transfer function.

Objectives:
- Load a simulated AGN continuum light curve
- Define a BLR transfer function
- Generate an emission-line light curve via convolution
- Recover the reverberation lag using both:
  - Cross-correlation analysis
  - Frequency-resolved lag analysis

## 1. Load the Continuum Light Curve

The continuum variability is generated from a power spectral density (PSD) model representative of AGN variability.

```python
import os

directory_path = r'C:\Users\Annie.LAPTOP-NJ5KV9M1\AGN_mapping\data'
file_path = os.path.join(directory_path, 'lc6hrs_365days.dat')

column_index1 = 0  
column_index2 = 1

with open(file_path, 'r') as file:
    # Read all lines
    lines = file.readlines()
    data_lines = lines[1:]
    
   # Extract the desired columns from each line and convert to float
    column1_data = [float(line.split()[column_index1]) for line in data_lines]
    column2_data = [float(line.split()[column_index2]) for line in data_lines]

time = np.array(column1_data)/(24*3600)
flux = np.array(column2_data)
```

## 2. Define the BLR Transfer Function

The BLR transfer function describes the delayed response of line emission to continuum variations.

The transfer function is normalised and characterised by:
- Total response
- Mean lag
- Effective width

```python
dt = 21600  # seconds corresponds to 6 hours
exposure = 1400  #length of array = days/6hours number of time bins
times = (np.linspace(1, exposure+1, num = 1400)*dt)/(24*3600) # converted to days equal to 1 year
#dt_days = 0.01   # instead of 0.25
#times = np.arange(1, 35000) * dt_days
S = 1
M = np.log(13.9)
#frequency = 2


BLR = 1/(S*times*np.sqrt(2*np.pi))*np.exp(-(np.log(times)-M)**2/(2*(S**2)))




```

## 3. Verification of the Transfer Function

Before generating reverberated light curves, we verify that the transfer function:

- Integrates to unity
- Produces the expected mean delay
- Has the intended shape

```python
#plot of BLR 
fig, ax = plt.subplots(1,1,figsize=(10,6))
ax.plot(times, BLR, lw=2, color='green')
#ax.set_xlim(-50,300)
ax.set_xlabel("Time (days)", fontproperties=font_prop)
ax.set_ylabel("Counts (cts)", fontproperties=font_prop)
ax.tick_params(axis='x', labelsize=16)
ax.tick_params(axis='y', labelsize=16)
plt.show()
```

```python
# integrate response function
normalisation = np.trapz(BLR, times)
mean_lag = np.trapz(BLR * times, times) 

print("Integral =", normalisation)
print(mean_lag)
```

## 4. Generate the Reverberated Light Curve

The emission-line light curve is generated by convolving the continuum light curve with the BLR transfer function:

\[
L(t)=\int C(t-\tau)\Psi(\tau)d\tau
\]

where:

- \(C(t)\) is the continuum light curve
- \(\Psi(\tau)\) is the BLR transfer function
- \(L(t)\) is the emission-line response

```python
# Perform convolution
signal4_BLR = np.convolve(flux, BLR, mode='full')
#signal4_BLR = signal4_BLR[100:1300]


zero_idx = np.argmin(np.abs(times))
signal4_BLR = signal4_BLR[zero_idx:zero_idx + len(flux)]

# ---- CROP EDGE EFFECTS ----

# Use BLR directly instead of psi_norm
threshold = 0.01 * np.max(BLR)
tf_width = np.max(times[BLR > threshold])

# Create mask to remove initial edge-affected region
mask = time > tf_width

# Apply mask to BOTH light curves
time_cut = time[mask]
flux_cut = flux[mask]
signal_cut = signal4_BLR[mask]

#print(signal_cut)
```

```python
#flux = flux[100:1300]
#time = time[100:1300]

lc4_BLR = Lightcurve(time_cut,signal_cut )
lc4 = Lightcurve(time_cut,flux_cut)

```

```python
from stingray.crosscorrelation import CrossCorrelation

cc = CrossCorrelation(lc4, lc4_BLR)

lags = cc.time_lags
ccf  = cc.corr

mask = ccf > 0.8 * np.max(ccf)

tau_cent = np.sum(
    lags[mask] * ccf[mask]
) / np.sum(ccf[mask])

print("Centroid of the CCF:", tau_cent)
```

## 8. Time-Domain Cross-Correlation Analysis

The cross-correlation function (CCF) provides an independent estimate of the reverberation lag.

The CCF centroid is expected to recover the mean delay of the transfer function.

```python
from stingray.crosscorrelation import CrossCorrelation

cc = CrossCorrelation(lc4_BLR_N_cut, lc4_N_cut)

lags = cc.time_lags
ccf  = cc.corr

mask = ccf > 0.8 * np.max(ccf)

tau_cent = np.sum(
    lags[mask] * ccf[mask]
) / np.sum(ccf[mask])

print("Centroid of the CCF:", tau_cent)
```
